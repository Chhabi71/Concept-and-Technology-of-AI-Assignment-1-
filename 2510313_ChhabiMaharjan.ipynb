6. Adding HDI Category Column:

```python
# HDI category classification
def classify_hdi(hdi):
    if hdi < 0.550:
        return "Low"
    elif 0.550 <= hdi <= 0.699:
        return "Medium"
    elif 0.700 <= hdi <= 0.799:
        return "High"
    else:
        return "Very High"

hdi_2022_df['HDI Category'] = hdi_2022_df['hdi'].apply(classify_hdi)
```

3. Outlier Detection:

```python
Q1 = hdi_southasia[['hdi', 'gross_inc_percap']].quantile(0.25)
Q3 = hdi_southasia[['hdi', 'gross_inc_percap']].quantile(0.75)
IQR = Q3 - Q1

outliers = hdi_southasia[
    (hdi_southasia['hdi'] < Q1['hdi'] - 1.5 * IQR['hdi']) |
    (hdi_southasia['hdi'] > Q3['hdi'] + 1.5 * IQR['hdi']) |
    (hdi_southasia['gross_inc_percap'] < Q1['gross_inc_percap'] - 1.5 * IQR['gross_inc_percap']) |
    (hdi_southasia['gross_inc_percap'] > Q3['gross_inc_percap'] + 1.5 * IQR['gross_inc_percap'])
]
```

```python
plt.figure()
sns.scatterplot(data=hdi_southasia, x='gross_inc_percap', y='hdi')
sns.scatterplot(data=outliers, x='gross_inc_percap', y='hdi', color='red')
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.title("Outlier Detection: HDI vs GNI")
plt.show()
```

7. Outlier Detection

```python
def detect_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    return outliers

# South Asia
South_Asia_out_hdi = detect_outliers(South_Asia_2020_22, 'hdi')
South_Asia_out_gni = detect_outliers(South_Asia_2020_22, 'gross_inc_percap')

# Middle East
Middle_East_out_hdi = detect_outliers(Middle_East_2020_22, 'hdi')
Middle_East_out_gni = detect_outliers(Middle_East_2020_22, 'gross_inc_percap')

# Scatter plots highlighting outliers
for region, df, out_df in zip(['South Asia','Middle East'],
                              [South_Asia_2020_22, Middle_East_2020_22],
                              [South_Asia_out_hdi, Middle_East_out_hdi]):
    plt.figure(figsize=(10,5))
    sns.scatterplot(x='gross_inc_percap', y='hdi', data=df, label='Normal')
    sns.scatterplot(x='gross_inc_percap', y='hdi', data=out_df, color='red', label='Outlier')
    plt.title(f"{region}: HDI vs GNI per Capita with Outliers")
    plt.xlabel("GNI per Capita")
    plt.ylabel("HDI")
    plt.legend()
    plt.show()
```

The scatter plot clearly indicates the outliners in the relationship between HDI and GNI per capita in the South Asia and the Middle East.
